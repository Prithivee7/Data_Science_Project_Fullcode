```python
import pandas as pd
import numpy as np
```

```python
job_data = pd.read_csv('Preprocessed Job Requirements.csv', low_memory=False)
```

```python
# Removing trailing spaces from all columns
for column_name in job_data.columns:
    column_list = job_data[column_name].tolist()
    for i in range(len(column_list)):
        if isinstance(column_list[i],str):
            column_list[i] = column_list[i].strip()
    job_data[column_name] = column_list
```

```python
resume_data = pd.read_csv('Preprocessed Resume Data2.csv')
```

```python
# Removing trailing spaces from all columns
for column_name in job_data.columns:
    column_list = job_data[column_name].tolist()
    for i in range(len(column_list)):
        if isinstance(column_list[i],str):
            column_list[i] = column_list[i].strip()
    job_data[column_name] = column_list  
```

```python
# Joining IT Skills, IT Skills Experience and Key Skills in resume data
IT_Skills = []
IT_Skills_Experience = []
Key_Skills = []
for i in range(4626):
    sample_it_skills_list = []
    sample_key_skills_list = []
    sample_it_skills_experience_list = []
    for j in range(1,31):
        var_it = resume_data['IT Skill '+str(j)].tolist()[i]
        if isinstance(var_it,str):
            sample_it_skills_list.append(var_it)
        var_key = resume_data['Key Skill '+str(j)].tolist()[i]
        if isinstance(var_key,str):
            sample_key_skills_list.append(var_key)
        var_experience = resume_data['IT Skill '+str(j)+' Experience'].tolist()[i]
        sample_it_skills_experience_list.append(str(var_experience))

    IT_Skills.append(','.join(sample_it_skills_list))
    Key_Skills.append(','.join(sample_key_skills_list))
    IT_Skills_Experience.append(','.join(sample_it_skills_experience_list))
resume_data['IT Skills'] = IT_Skills
resume_data['Key Skills'] = Key_Skills
resume_data['IT Skills Experience'] = IT_Skills_Experience
```

```python
# Joining IT Skills and Key Skills in job requirements data
IT_Skills = []
Key_Skills = []
for i in range(5090):
    sample_it_skills_list = []
    sample_key_skills_list = []
    for j in range(1,255):
        var_it = job_data['IT Skill '+str(j)].tolist()[i]
        if isinstance(var_it,str):
            sample_it_skills_list.append(var_it)
    for j in range(1,37):    
        var_key = job_data['Key Skills '+str(j)].tolist()[i]
        if isinstance(var_key,str):
            sample_key_skills_list.append(var_key)

    IT_Skills.append(','.join(sample_it_skills_list))
    Key_Skills.append(','.join(sample_key_skills_list))
    
job_data['IT Skills'] = IT_Skills
job_data['Key Skills'] = Key_Skills
```

```python
# Locations from job requirements data
locations = set()
for i in range(1,9):
    for location in job_data['Location '+str(i)]:
        if isinstance(location,str):
            locations.add(location.lower())
#locations
```

```python
# Changing names of the cities which has multiple names in job requirements data
for i in range(1,9):
    locations = job_data['Location '+str(i)].tolist()
    for num in range(len(locations)):
        if isinstance(locations[num],str):
            if locations[num].lower() == 'bangalore':
                locations[num] = 'Bengaluru'
            elif locations[num].lower() == 'gurgaon':
                locations[num] = 'Gurugram'
            elif locations[num].lower() == 'trivandrum':
                locations[num] = 'thiruvananthapuram'
            elif locations[num].lower() == 'cochin':
                locations[num] = 'Kochi'
            elif locations[num].lower() == 'pondicherry':
                locations[num] = 'Puducherry'
            elif locations[num].lower() == 'mangalore':
                locations[num] = 'Mangaluru'
    job_data['Location '+str(i)] = locations
            
```

```python
# Locations from resume data
locations = set()
for i in range(1,4):
    for location in resume_data['Preferred Location '+str(i)]:
        if isinstance(location,str):
            locations.add(location.lower())
# locations
```

```python
# Changing names of the cities which has multiple names in resume data
for i in range(1,4):
    locations = resume_data['Preferred Location '+str(i)].tolist()
    for num in range(len(locations)):
        if isinstance(locations[num],str):
            if locations[num].lower() == 'bangalore':
                locations[num] = 'Bengaluru'
            elif locations[num].lower() == 'gurgaon':
                locations[num] = 'Gurugram'
            elif locations[num].lower() == 'trivandrum':
                locations[num] = 'thiruvananthapuram'
            elif locations[num].lower() == 'cochin':
                locations[num] = 'Kochi'
            elif locations[num].lower() == 'pondicherry':
                locations[num] = 'Puducherry'
            elif locations[num].lower() == 'mangalore':
                locations[num] = 'Mangaluru'
    resume_data['Preferred Location '+str(i)] = locations
```

```python
job_data.to_csv('Job Requirements.csv', index=False)
resume_data.to_csv('Resume.csv', index=False)
```

```python
# No. of IT Skills in resume data = 30
# No. of Key Skills in resume data = 30
# No. of IT Skills in job requirements data = 254
# No. of Key Skills in job requirements data = 36
# Total number of rows in resume data = 4626
# Total number of rows in job requirements data = 5090
```

```python
# Recommend Job Requirements for a given resume
index = 420
# resume_data.iloc[[index]]
input_resume = resume_data.iloc[[index]]
IT_Skills = [x.lower() for x in input_resume['IT Skills'].tolist()[0].split(',')]
IT_Skills_Experience = [float(x) for x in input_resume['IT Skills Experience'].tolist()[0].split(',')]
Key_Skills = [x.lower() for x in input_resume['Key Skills'].tolist()[0].split(',')]

Total_Score = []
for num in range(5090):
    score = 0
    job = job_data.iloc[[num]]
    
    Job_IT_Skills = [x.lower() for x in job['IT Skills'].tolist()[0].split(',')]
    Job_Key_Skills = [x.lower() for x in job['Key Skills'].tolist()[0].split(',')]

    for i in range(len(IT_Skills)):
        if IT_Skills[i] in Job_IT_Skills:
            score = score + (IT_Skills_Experience[i] * 1 + 1) * (Job_IT_Skills.count(IT_Skills[i]))
    for i in range(len(Key_Skills)):
        if Key_Skills[i] in Job_Key_Skills:
            score = score + 1
    Total_Score.append(score)
# print(Total_Score)
output = np.argsort(np.array(Total_Score))[::-1]
# input_resume.to_csv('Input(Resume).csv')
job_data.iloc[output[0:10]]
```
