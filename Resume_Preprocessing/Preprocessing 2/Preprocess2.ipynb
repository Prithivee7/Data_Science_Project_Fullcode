```python
import pandas as pd
```

```python
data = pd.read_csv('Preprocessed Job Requirements1.csv', low_memory=False)
```

```python
ug_data = data['UG'].tolist()

for i in range(len(ug_data)):
    if isinstance(ug_data[i],str):
        ug_data[i] = ','.join(ug_data[i].split(',')[0:8])

data['UG'] = ug_data
```

```python
name_list = []
for i in range(1,9):
    name_list.append('UG '+str(i))

data[name_list] = data['UG'].str.split(",", expand = True)
```

```python
branches = {}
branches['Accounting'] = ['Accounting', 'Account']
branches['Aerospace'] = ['Aerospace']
branches['Agriculture'] = ['Agriculture']
branches['Arts'] = ['Arts', 'Applied Arts', 'B.A in Any Specialization', 'B.A in Other Specialization']
branches['Electrical, Electronics & Communication'] = ['Electronic', 'Electrical', 'E.C.E', 'ECE', 'EEE', 'Communication', 'ENTC', 'I and E', 'Instrumentation']
branches['Automobile'] = ['Automobile']
branches['Political Science'] = ['Politics', 'Political Science']
branches['Industry'] = ['ISE', 'Industry', 'Industrial', 'Production']
branches['Biology'] = ['Bio']
branches['Computers'] = ['C.S.E', 'Computer', 'CSE', 'IT', 'Java', 'HTML', 'SQL', 'Network', 'Software', 'I.T', 'Information', 'CS', 'Data', 'Web Design', 'BCA in Any Specialization']
branches['Chemistry'] = ['Chemistry', 'Chemical']
branches['Civil'] = ['Civil']
branches['Commerce'] = ['Commerce', 'B.Com']
branches['Mathematics'] = ['Mathematics', 'Maths']
branches['Statistics'] = ['Statistics']
branches['Economics'] = ['Economic']
branches['Education'] = ['Education', 'B.El.Ed']
branches['English'] = ['English']
branches['Finance'] = ['Finance']
branches['Geology'] = ['Geology', 'Geography']
branches['History'] = ['History']
branches['Hotel Management'] = ['Hotel']
branches['Hindi'] = ['Hindi']
branches['Journalism'] = ['Journalism']
branches['Management'] = ['HR', 'Management', 'Administration']
branches['Science'] = ['B.Z.C', 'Botany', 'Physics', 'Science', 'Zoology', 'B.Sc in Any Specialization']
branches['Marketing'] = ['Marketing']
branches['Mechanical'] = ['Mechanical', 'Robotic', 'Mechatronics', 'Robotics']
branches['Metallurgy'] = ['Metallurgy']
branches['Mining'] = ['Mining']
branches['Petroleum'] = ['Petrol']
branches['Pharmacy'] = ['Pharmacy']
branches['Plastics'] = ['Plastics']
branches['Psychology'] = ['Psychology']
branches['Technology'] = ['B.Tech/B.E. in Any Specialization']
branches['Sociology'] = ['Social']
branches['Law'] = ['Law']
branches['Energy'] = ['Energy']
branches['Architecture'] = ['Architecture']
branches['Other Fields'] = ['Other']
branches['Any Field'] = ['Any Graduate', 'Any Specialization']
branches['Not Required'] = ['Not Required']

for i in range(1,9):
    ug_list = []
    for j in data['UG '+str(i)]:
        if isinstance(j, str):
            flag = 0
            for key in branches.keys():
                for value in branches[key]:
                    if value.lower() in j.lower():
                        flag = 1
                        if key == 'Not Required':
                            ug_list.append('Degree ' + key)
                            break
                        if 'Diploma'.lower() in j.lower():
                            ug_list.append('Diploma in ' + key)
                        else:
                            ug_list.append('Degree in ' + key)
                        break
                if flag == 1:
                    break
            if flag == 0:
                ug_list.append('Degree in Other Fields')
        else:
            ug_list.append(j)
    data['UG ' + str(i)] = ug_list
```

```python
doctorate = data['DOCTORATE'].tolist()

for i in range(len(doctorate)):
    if isinstance(doctorate[i],str):
        doctorate[i] = ','.join(doctorate[i].split(',')[0:1])

doctorate_list = []
for i in doctorate:
    if isinstance(i, str):
        flag = 0
        for key in branches.keys():
            for value in branches[key]:
                if value.lower() in i.lower():
                    flag = 1
                    if key == 'Not Required':
                        doctorate_list.append('Doctorate ' + key)
                        break
                    doctorate_list.append('Doctorate in ' + key)
                    break
            if flag == 1:
                break
        if flag == 0:
            doctorate_list.append('Doctorate in Other Fields')
    else:
        doctorate_list.append(i)

data['DOCTORATE'] = doctorate_list
```
