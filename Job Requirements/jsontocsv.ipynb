```python
import json 
import csv 
import os
```

```python
job_descriptions = open('job_descriptions.csv', 'w', newline='') 
csv_writer = csv.writer(job_descriptions)
count = 0
for file in os.listdir('./software-developer'):
    path = './software-developer/{0}'.format(file)
    print(path)
    with open(path) as f:
        data = json.load(f)
    if count == 0:
        header = data.keys() 
        csv_writer.writerow(header) 
        count += 1
    csv_writer.writerow(data.values())
job_descriptions.close() 
```
