```python
import json 
import csv 
import os
```

```python
job_descriptions = open('Job Requirements.csv', 'w', newline='')
csv_writer = csv.writer(job_descriptions)
count = 0
header = ['Job Title','Company','Experience','Salary','Location','Job Description','Role','Industry Type','Functional Area','Employment Type','Role Category','UG','PG','Doctorate','Key Skills','Link']
for num in range(1, len(os.listdir('./Updated_Job_Requirements'))+1):
    path = './Updated_Job_Requirements/'+str(num)+".json"
    print(path)
    with open(path) as f:
        data = json.load(f)
    if count == 0: 
        csv_writer.writerow(header) 
        count += 1
    for i in data.keys():
        if u'\u20b9' in data[i]:
            data[i] = data[i].replace(u'\u20b9','')
    values = []
    for key in header:
        if key in data:
            values.append(data[key])
        else:
            values.append("None")
    csv_writer.writerow(values)
job_descriptions.close()
```
